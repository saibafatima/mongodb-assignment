```python
Q2. State and Explain the features of MongoDB.

Some key features of MongoDB include:

Document-based data model: MongoDB stores data in JSON-like documents, which can contain various data types.
Schema flexibility: MongoDB allows for dynamic schema changes, making it ideal for applications with evolving data structures.
Scalability: MongoDB is designed for horizontal scaling, making it easy to add more nodes to handle increasing data volumes and traffic.
High performance: MongoDB uses an in-memory data storage engine, which provides fast data retrieval and query execution.
Rich query language: MongoDB provides a powerful query language, allowing for complex queries and data retrieval.
Data replication: MongoDB supports data replication, ensuring high availability and fault tolerance.
```

```python
#Q4. Using the database and the collection created in question number 3, write a code to insert one record, and insert many records. Use the find() and find_one() methods to print the inserted record.

from pymongo import MongoClient

# Step 1: Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')
db = client['mydatabase']
collection = db['mycollection']

# Step 2: Insert one record
single_record = {
    "name": "John Doe",
    "age": 30,
    "email": "johndoe@example.com"
}
inserted_single_record_id = collection.insert_one(single_record).inserted_id
print(f"Inserted one record with ID: {inserted_single_record_id}")

# Step 3: Insert many records
multiple_records = [
    {
        "name": "Jane Smith",
        "age": 25,
        "email": "janesmith@example.com"
    },
    {
        "name": "Bob Johnson",
        "age": 35,
        "email": "bobjohnson@example.com"
    },
    {
        "name": "Alice Williams",
        "age": 28,
        "email": "alicewilliams@example.com"
    }
]
inserted_multiple_records_ids = collection.insert_many(multiple_records).inserted_ids
print(f"Inserted multiple records with IDs: {inserted_multiple_records_ids}")

# Step 4: Use find_one() to print the inserted single record
found_single_record = collection.find_one({"_id": inserted_single_record_id})
print("Found single record:")
print(found_single_record)

# Step 5: Use find() to print all inserted records
print("Found all records:")
for record in collection.find():
    print(record)
```

```python
#Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.

from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')
db = client['mydatabase']
collection = db['mycollection']

# Insert some sample data
sample_data = [
    {"name": "John Doe", "age": 30, "email": "johndoe@example.com"},
    {"name": "Jane Smith", "age": 25, "email": "janesmith@example.com"},
    {"name": "Bob Johnson", "age": 35, "email": "bobjohnson@example.com"},
    {"name": "Alice Williams", "age": 28, "email": "alicewilliams@example.com"}
]
collection.insert_many(sample_data)

# 1. Basic Query - Retrieve all documents
print("All documents:")
for doc in collection.find():
    print(doc)

# 2. Query with criteria - Retrieve documents where age is greater than 28
print("\nDocuments where age is greater than 28:")
query = {"age": {"$gt": 28}}
for doc in collection.find(query):
    print(doc)

# 3. Projection - Retrieve only the name and age fields
print("\nDocuments with only name and age fields:")
projection = {"_id": 0, "name": 1, "age": 1}
for doc in collection.find({}, projection):
    print(doc)

# 4. Sorting - Retrieve documents sorted by age in descending order
print("\nDocuments sorted by age (descending):")
for doc in collection.find().sort("age", -1):
    print(doc)

# 5. Limit - Retrieve only the first 2 documents
print("\nFirst 2 documents:")
for doc in collection.find().limit(2):
    print(doc)

# 6. Skip - Skip the first 2 documents and retrieve the rest
print("\nDocuments after skipping the first 2:")
for doc in collection.find().skip(2):
    print(doc)
```
